```python
import pandas as pd
import json


input_filepath = 'D:\LLM-Hallucination\data\SVAMP\SVAMP.json'

# 2. The path to your output file
output_filepath = 'D:\LLM-Hallucination\data\prompts\svamp_prompts.json'

# 3. The number of rows to sample
num_samples_to_add = 200

# ---------------------

def sample_and_append(input_file, output_file, num_samples):
    
    print(f"Loading local data from '{input_file}'...")
    try:
        df = pd.read_json(input_file)
    except FileNotFoundError:
        print(f"Error: The file '{input_file}' was not found. Make sure it's in the same directory.")
        return

    
    df = df.rename(columns={
        'Body': 'context',
        'Question': 'question',
        'ID': 'id',
        'Answer': 'answer_text'
    })
    
    
    print(f"Found {len(df)} total questions.")
    if len(df) < num_samples:
        sampled_df = df
    else:
        print(f"Randomly sampling {num_samples} questions...")
        sampled_df = df.sample(n=num_samples, random_state=42)


    output_columns = ['id', 'context', 'question', 'answer_text']
    final_samples_df = sampled_df[output_columns]
    
    samples_to_append = final_samples_df.to_dict(orient='records')

    
    try:
        with open(output_file, 'r', encoding='utf-8') as f:
            existing_data = json.load(f)
    except (FileNotFoundError, json.JSONDecodeError):
        existing_data = []

    existing_data.extend(samples_to_append)

    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(existing_data, f, indent=4)

    print("-" * 20)
    print("Process complete!")
    print(f"The file '{output_file}' now contains a total of {len(existing_data)} entries.")


sample_and_append(input_filepath, output_filepath, num_samples_to_add)


```
